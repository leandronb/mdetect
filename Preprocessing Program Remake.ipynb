```python
#%plotly inline
import plotly.plotly as py
import pandas as pd
from scipy.signal import medfilt, butter, lfilter, freqz
import matplotlib
import cufflinks as cf
import numpy as np


# Set Directory path and file name
RAW_DIR = "HAPT/RawData/"
RAW_FILE = "acc_exp01_user01.txt"


# Butterworth Filters  
def butter_lowpass(cutoff, fs, order=3):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_highpass(cutoff, fs, order=3):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=3):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

def butter_highpass_filter(data, cutoff, fs, order=3):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y


#Plot Configuraton Helper
def plot_config(df):
    config = [{
        'x': df.index,
        'y': df[col],
        'name': col
    }  for col in df.columns]
    return config


# Read Data
df = pd.read_csv(RAW_DIR + RAW_FILE, header = None, delim_whitespace = True)

# Apply median filter to the data
mfiltered = medfilt(df)

# Apply Butterworth lowpass filter
blfiltered = butter_lowpass_filter(mfiltered, 20, 50, 3)
# Apply Butterworth highpass filter to remove gravity
bhfiltered = butter_highpass_filter(mfiltered, 0.3, 50, 3)

# Convert data into a dataframes
#blfiltered_df = pd.DataFrame(blfiltered)
bhfiltered_df = pd.DataFrame(bhfiltered)

bhfiltered_df[0]





```
